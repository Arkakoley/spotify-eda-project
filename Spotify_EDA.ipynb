**Missing values & duplicates**

```python
# missing values
print(df.isnull().sum().sort_values(ascending=False).head(20))

# duplicates
print("duplicates:", df.duplicated().sum())
df = df.drop_duplicates()
```

**Convert important columns to numeric (safe coercion)**

```python
num_cols = ['popularity','danceability','energy','acousticness','duration_ms',
            'instrumentalness','liveness','loudness','speechiness','tempo','valence']

for c in num_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors='coerce')
```

**Univariate plots (examples)**

```python


# Duration boxplot (seconds)
plt.figure(figsize=(6,3))
sns.boxplot(x=(df['duration_ms']/1000).dropna())
plt.xlabel('Duration (sec)')
plt.show()

# Energy distribution
plt.figure(figsize=(8,4))
sns.histplot(df['energy'].dropna(), kde=True, color='orange')
plt.title('Energy Distribution')
plt.xlabel('Energy')
plt.ylabel('Count')
plt.show()

# Danceability distribution
plt.figure(figsize=(8,4))
sns.histplot(df['danceability'].dropna(), kde=True, color='limegreen')
plt.title('Danceability Distribution')
plt.xlabel('Danceability')
plt.ylabel('Count')
plt.show()
```

```python
# Popularity distribution
plt.figure(figsize=(8,4))
sns.histplot(df['popularity'].dropna(), kde=True)
plt.title('Popularity distribution')
plt.show()
```

**Correlation heatmap**

```python
plt.figure(figsize=(10,8))
corr = df[num_cols].corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', linewidths=.5)
plt.title('Correlation matrix')
plt.show()
```

```python
# Boxplot for Duration (in seconds)
plt.figure(figsize=(6,4))
sns.boxplot(x=df['duration_ms'] / 1000, color='skyblue')
plt.title('Boxplot of Song Duration')
plt.xlabel('Duration (seconds)')
plt.show()

# Boxplot for Loudness
plt.figure(figsize=(6,4))
sns.boxplot(x=df['loudness'], color='salmon')
plt.title('Boxplot of Loudness')
plt.xlabel('Loudness (dB)')
plt.show()
```

```python
# Countplot for Language
plt.figure(figsize=(8,4))
sns.countplot(x=df['language'], palette='pastel')
plt.title('Countplot of Language')
plt.xlabel('Language')
plt.ylabel('Count')
plt.show()

# Countplot for Mode
plt.figure(figsize=(6,4))
sns.countplot(x=df['mode'], palette='muted')
plt.title('Countplot of Mode')
plt.xlabel('Mode')
plt.ylabel('Count')
plt.show()


```

```python
# Countplot for Key
plt.figure(figsize=(8,4))
sns.countplot(x=df['key'], palette='bright')
plt.title('Countplot of Key')
plt.xlabel('Key')
plt.ylabel('Count')
plt.show()

# Countplot for Time Signature
plt.figure(figsize=(6,4))
sns.countplot(x=df['time_signature'], palette='dark')
plt.title('Countplot of Time Signature')
plt.xlabel('Time Signature')
plt.ylabel('Count')
plt.show()
```

**Bivariate example (danceability vs popularity)**

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='danceability', y='popularity', alpha=0.6)
plt.title('Danceability vs Popularity')
plt.show()
```

```python
# Scatterplot: Energy vs Popularity
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='energy', y='popularity', alpha=0.6, color='green')
plt.title('Energy vs Popularity')
plt.show()
```

```python

# Scatterplot: Loudness vs Energy
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='loudness', y='energy', alpha=0.6, color='red')
plt.title('Loudness vs Energy')
plt.show()
```

**Time-series trend by year**

```python
if 'year' in df.columns:
    year_trend = df.groupby('year')['popularity'].mean().reset_index()
    plt.figure(figsize=(10,4))
    sns.lineplot(data=year_trend, x='year', y='popularity', marker='o')
    plt.title('Average popularity by year')
    plt.show()
```

```python
if 'year' in df.columns:
    # Energy over time
    energy_trend = df.groupby('year')['energy'].mean().reset_index()
    plt.figure(figsize=(10,4))
    sns.lineplot(data=energy_trend, x='year', y='energy', marker='o')
    plt.title('Average Energy by Year')
    plt.show()


```

```python
   if 'year' in df.columns: # Loudness over time
    loudness_trend = df.groupby('year')['loudness'].mean().reset_index()
    plt.figure(figsize=(10,4))
    sns.lineplot(data=loudness_trend, x='year', y='loudness', marker='o')
    plt.title('Average Loudness by Year')
    plt.show()

```

```python
if 'year' in df.columns: # Valence over time
    valence_trend = df.groupby('year')['valence'].mean().reset_index()
    plt.figure(figsize=(10,4))
    sns.lineplot(data=valence_trend, x='year', y='valence', marker='o')
    plt.title('Average Valence by Year')
    plt.show()
```

# Spotify Tracks EDA Project ðŸŽµ

## Overview
This project performs an **Exploratory Data Analysis (EDA)** on a dataset of Spotify tracks.  
The goal is to understand how different audio features (like danceability, energy, loudness, tempo, valence, etc.) relate to a trackâ€™s **popularity** and how these characteristics have changed over time.

## Key Analyses Performed
- **Univariate analysis:** distributions of popularity, danceability, energy, etc.  
- **Bivariate analysis:** relationships such as danceability vs popularity, energy vs valence.  
- **Correlation heatmap:** to identify strongest relationships between features.  
- **Time-series trends:** how song characteristics evolved over years.  
- **Recommendations:** insights for music producers or mixing engineers.

## Tools Used
- Python 3  
- pandas, numpy, matplotlib, seaborn, scikit-learn  

## Insights Summary
- Highly popular songs tend to have **high danceability & energy**.  
- **Average loudness has increased** while **song duration decreased** over the years.  
- Songs in multiple languages (like Spanish & Korean) show rising popularity trends.
